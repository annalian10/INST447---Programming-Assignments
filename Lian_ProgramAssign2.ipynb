```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('InstacartOrdersByDepartment.csv')
```

```python
df.shape
```

### Research Question: _Do people order different foods at different times of day?_

```python
alcohol = df[df['department'] == 'alcohol']
plt.figure(figsize=(10,6))
alcohol.plot.bar(x='order_hour_of_day', y='num_orders_hour')

plt.title('Alcohol Orders by Hour of the Day', fontsize = 16)
plt.xlabel('Hour of Day', fontsize = 12)
plt.ylabel('Number of Orders', fontsize = 12)
```

This graph of alcohol orders by hour of the day shows that there's a peak of orders in the evening especially around 3pm, with relatively low activity during the early morning hours. This makes sense because people tend to drink alcohol during social events or after work, which usually occurs in the evening. This bar pattern answers the research question by showing that people are more likely to order alcohol later in the day.

```python

snacks = df[df['department'] == 'snacks']
plt.figure(figsize=(10,6))
snacks.plot.bar(x='order_hour_of_day', y='num_orders_hour')

plt.title('Snacks Orders by Hour of the Day', fontsize = 16)
plt.xlabel('Hour of Day', fontsize = 12)
plt.ylabel('Number of Orders', fontsize = 12)
```

Snacks orders increase as the day goes on, peaking in the late afternoon and evening, from 2pm to 8pm. This suggests that people are more likely to order snacks when they feel hungry later in the day, perhaps after work or school. Snacks orders are lower late at night (11pm - 3am). This pattern shows that people tend to crave and order snacks in the second half of the day. 

```python
produce = df[df['department'] == 'produce']
plt.figure(figsize=(10,6))
produce.plot.bar(x='order_hour_of_day', y='num_orders_hour')

plt.title('Produce Orders by Hour of the Day', fontsize = 16)
plt.xlabel('Hour of Day', fontsize = 12)
plt.ylabel('Number of Orders', fontsize = 12)
```

In this graph, it shows that there's a clear trend of higher orders during the morning and early afternoon hours (8am and to 2pm), which makes sense because people might want to buy fresh fruits and vegetables early in the day if they're planning their meals. After 4pm, produce orders drop off, showing that people usually plan their fresh food purchases earlier in the day. 

#### Conclusion (150 words minimum)

Across the three visualizations, I observed distinct ordering patters in the Alcohol, Produce, and Snacks departments based on the time of day. **Alcohol orders** tend to peak during the evening hours, aligning with people that buying alcohol for social events or to unwind after the day. **Produce orders** are highest in the morning and early afternoon, suggesting that people are likely planning meals for the week or grocery shopping for the day. Meanwhile **Snacks orders** show a steady increase throughout the day, with peaks in the late afternoon and evening, likely driven by impulse buys or cravings.
